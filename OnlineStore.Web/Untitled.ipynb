```python
import pandas as pd
import numpy as np
import scipy 
import random
from collections import Counter
import math

d = pd.read_csv("Products.csv")

product_price = []

discounted_price = []

final_price = []

for x in d["Price"]:
    product_price.append(x)

for x in d["DiscountedPrice"]:
    discounted_price.append(x)

for x in range(len(product_price)):
    final_price.append(product_price[x] - discounted_price[x])


# id_list - selected_customer(X) - unique_order_id - sum_list - selected_address(X) - is_delivered
```

```python

# print(final_price) # final price of each product

e = pd.read_csv("OrderedProducts.csv")

# ordered products

order_id = []

for x in e["OrderId"]:
    order_id.append(x)

product_id = []

for x in e["ProductId"]:
    product_id.append(x)

quantity = []

for x in e["Quantity"]:
    quantity.append(x)


net_price_list = []

for a in range(len(order_id)):
    net_price_list.append(final_price[product_id[a] - 1] * quantity[a])


ordered_products = [order_id, net_price_list]

```

```python

def bring_sum(example_id):
    sum = 0
    for a in range(len(order_id)):
        if(ordered_products[0][a] == example_id):
            sum = sum + ordered_products[1][a]
        else:
            continue
    return sum


unique_order_id = []

for x in order_id:
    if x not in unique_order_id:
        unique_order_id.append(x)

sum_list = []

for x in unique_order_id:
    sum_list.append(bring_sum(x))


order_id_and_total_price_final = [unique_order_id, sum_list]

id_list = []

for i in range(1,614):
    id_list.append(i)

```
